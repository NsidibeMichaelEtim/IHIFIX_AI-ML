The co-pilot loop is how a human and an AI take turns working together on a task.
Here is how it works
1. You give the AI a goal — a task, question, or instruction.
2. The AI makes a plan — it figures out what steps to take and what tools it needs.
3. The AI acts — writes code, searches, edits a file, whatever the task needs.
4. It checks the results — did it work? Did something go wrong?
5. It reports back to you — shows what it did, what it found, or asks a question.
6. You review and respond — you approve, correct, or give the next instruction. Then the whole thing starts again.

```python
# Import libraries
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv("titanic_data.csv")
```

Basic Exploration

```python
# View first 5 rows
print("FIRST 5 ROWS")
print(titanic_data.head())

# Dataset information
print("\nDATASET INFO")
print(titanic_data.info())

# Statistical summary
print("\nSTATISTICAL SUMMARY")
print(titanic_data.describe())
```

Solution to Analysis Questions

```python
# 1. How many passengers survived?
survived_count = titanic_data['Survived'].sum()

print("\n1. NUMBER OF PASSENGERS WHO SURVIVED:")
print(survived_count)

print("Why this makes sense:")
print("The 'Survived' column uses 1 for survived and 0 for not survived, so summing gives the total survivors.")
```

```python
# 2. What was the average age?
average_age = titanic_data['Age'].mean()

print("\n2. AVERAGE AGE OF PASSENGERS:")
print(average_age)

print("Why this makes sense:")
print("The mean() function calculates the average of all available passenger ages.")
```

```python
# 3. Show only first-class women
first_class_women = titanic_data[
    (titanic_data['Pclass'] == 1) &
    (titanic_data['Sex'] == 'female')
]

print("\n3. FIRST-CLASS WOMEN:")
print(first_class_women)

print("Why this makes sense:")
print("The filter keeps only passengers who were in first class and female.")
```

```python
# 4. Group by class and show survival rate
survival_rate = titanic_data.groupby('Pclass')['Survived'].mean()

print("\n4. SURVIVAL RATE BY PASSENGER CLASS:")
print(survival_rate)

print("Why this makes sense:")
print("Taking the average of the Survived column gives the proportion of survivors in each class.")
```

```python
# 5. Drop rows with missing age — how many rows removed?
original_rows = titanic_data.shape[0]

cleaned_data = titanic_data.dropna(subset=['Age'])

new_rows = cleaned_data.shape[0]

rows_removed = original_rows - new_rows

print("\n5. ROWS REMOVED AFTER DROPPING MISSING AGES:")
print(rows_removed)

print("Why this makes sense:")
print("Rows with missing Age values are removed because age-related analysis requires complete age information.")
```

TASK 3 — Five chart types (Titanic)
On the same dataset, create five charts: histogram, scatter, bar, box plot, line chart. Each chart must have a title and axis labels. Re-create all five using seaborn styling and note which version you prefer.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titanic_data = pd.read_csv("titanic_data.csv")
```

CHARTS USING MATPLOTLIB

```python
# 1. Histogram
plt.figure(figsize=(7,5))
plt.hist(titanic_data['Age'].dropna(), bins=20)
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Scatter Plot
plt.figure(figsize=(7,5))
plt.scatter(titanic_data['Age'], titanic_data['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# 3. Bar Chart
class_survival = titanic_data.groupby('Pclass')['Survived'].sum()

plt.figure(figsize=(7,5))
plt.bar(class_survival.index, class_survival.values)
plt.title("Number of Survivors by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Survivors")
plt.show()
```

```python
# 4. Box Plot
plt.figure(figsize=(7,5))
plt.boxplot(titanic_data['Age'].dropna())
plt.title("Box Plot of Passenger Ages")
plt.ylabel("Age")
plt.show()
```

```python
# 5. Line Chart
age_by_class = titanic_data.groupby('Pclass')['Age'].mean()

plt.figure(figsize=(7,5))
plt.plot(age_by_class.index, age_by_class.values, marker='o')
plt.title("Average Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
salary_data = pd.read_csv("ds_salaries.csv")

# View dataset
print(salary_data.head())

# Dataset info
print(salary_data.info())
```

QUESTION 1
Which experience level earns the highest average salary?

```python
# Average salary by experience level
experience_salary = salary_data.groupby('experience_level')['salary_in_usd'].mean()

# Plot
plt.figure(figsize=(8,5))

sns.barplot(
    x=experience_salary.index,
    y=experience_salary.values
)

plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary in USD")

plt.show()

print(experience_salary)
```

Finding Sentence

```python
print("Finding:")
print("Executive-level employees earned the highest average salary, showing that experience strongly affects income.")
```

QUESTION 2
Which company size pays the highest salaries?


```python
# Average salary by company size
company_salary = salary_data.groupby('company_size')['salary_in_usd'].mean()

# Plot
plt.figure(figsize=(8,5))

sns.barplot(
    x=company_salary.index,
    y=company_salary.values
)

plt.title("Average Salary by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Average Salary in USD")

plt.show()

print(company_salary)
```

Finding Sentence

```python
print("Finding:")
print("Large companies generally paid the highest average salaries compared to small and medium companies.")
```

QUESTION 3
Does remote work affect salary?

```python
# Average salary by remote ratio
remote_salary = salary_data.groupby('remote_ratio')['salary_in_usd'].mean()

# Plot
plt.figure(figsize=(8,5))

sns.lineplot(
    x=remote_salary.index,
    y=remote_salary.values,
    marker='o'
)

plt.title("Average Salary by Remote Work Ratio")
plt.xlabel("Remote Ratio")
plt.ylabel("Average Salary in USD")

plt.show()

print(remote_salary)
```

Finding Sentence

```python
print("Finding:")
print("Fully remote workers often earned higher average salaries than workers with no remote flexibility.")
```
